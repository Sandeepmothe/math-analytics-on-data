```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("income1.csv", names=["income", "count"], skiprows=[0])
df.head()
```

```python
sns.barplot(x = "income", y= "count", data = df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x= "income", y= "count", data= df)
plt.xscale("log")
plt.show()
```
